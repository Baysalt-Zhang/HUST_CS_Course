```python
'''
用改进的欧拉法求解如下初值问题
'''
import matplotlib.pyplot as plt

h = 0.2  # 步长

yy = []
xx = []

yy.append(1)
xx.append(0)


def f(m, n):
    return m + n


for i in range(6):
    k1 = f(yy[i], xx[i])
    y = yy[i] + h * f(yy[i], xx[i])
    xx.append(xx[i] + h)
    k2 = f(y, xx[i + 1])
    yy.append(yy[i] + h * 1.0 / 2.0 * (k2 + k1))
    print("y = ", yy[i], "x = %.1f" % xx[i])

plt.plot(xx, yy)
plt.show()


```

```python
import matplotlib.pyplot as plt

h = 0.2  # 步长

y_list = []
x_list = []

y_list.append(1)
x_list.append(0)


def f(m, n):
    return m + n


for i in range(6):
    k1 = h * f(x_list[i], y_list[i])
    k2 = h * f(x_list[i] + 1.0/2.0*h, y_list[i] + 1.0/2.0*k1)
    k3 = h * f(x_list[i] + 1.0/2.0*h, y_list[i] + 1.0/2.0*k2)
    k4 = h * f(x_list[i] + h, y_list[i] + k3)
    x_list.append(x_list[i] + h)
    y_list.append(y_list[i] + 1.0/6.0*(k1+2*k2+2*k3+k4))
    print("y = ", y_list[i], "x = %.1f" % x_list[i])

plt.plot(x_list, y_list)
plt.show()

```
